```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
image = cv2.imread('/content/drive/MyDrive/IPCV/IPCV1/rose.jpg', 1) # cv2 read image as BGR, 0 -> gray, 1(defaut) -> color
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # convert to RGB
```

```python
type(image)     # type image as numpy array
```

```python
print(image.shape)
if len(image.shape) == 2:
  h, w = image.shape                           # height, width, channel = None if image gray
else:
  h, w, c = image.shape                        # height, width, channel = 3 if image color
```

```python
def show_image(image, title):
  plt.figure()
  plt.imshow(image)
  plt.axis("off")
  plt.title(title)
  plt.show()
```

```python
show_image(image, "Original")
```

```python
def show_channel():
  image_b = image.copy()
  image_g = image.copy()
  image_r = image.copy()

  image_b[:, :, (0,1)] = 0    # assign channel 0(Red) and channel 1(Green) to 0
  image_g[:, :, (0,2)] = 0    # assign channel 0(Red) and channel 2(Blue) to 0
  image_r[:, :, (1,2)] = 0    # assign channel 1(Green) and channel 2(Blue) to 0

  show_image(image_b, "Blue Channel")   # current image just Blue
  show_image(image_g, "Green Channel")  # current image just Green
  show_image(image_r, "Red Channel")    # current image just Red
```

```python
def Show_LeftorRight_Gray(choose):
  if choose == "L":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    image_cvt[:, :w//2, 0] = image_cvtGray[:, :w//2]
    image_cvt[:, :w//2, 1] = image_cvtGray[:, :w//2]
    image_cvt[:, :w//2, 2] = image_cvtGray[:, :w//2]
    show_image(image_cvt, "Left Width Gray and Right Width Color")
  elif choose == "R":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    image_cvt[:, w//2:, 0] = image_cvtGray[:, w//2:]
    image_cvt[:, w//2:, 1] = image_cvtGray[:, w//2:]
    image_cvt[:, w//2:, 2] = image_cvtGray[:, w//2:]
    show_image(image_cvt, "Left Width Color and Right Width Gray")
  elif choose == "LR":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    image_cvt[:, :w//2, 0] = image_cvtGray[:, :w//2]
    image_cvt[:, :w//2, 1] = image_cvtGray[:, :w//2]
    image_cvt[:, :w//2, 2] = image_cvtGray[:, :w//2]
    show_image(image_cvt, "Left Width Gray and Right Width Color")
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    image_cvt[:, w//2:, 0] = image_cvtGray[:, w//2:]
    image_cvt[:, w//2:, 1] = image_cvtGray[:, w//2:]
    image_cvt[:, w//2:, 2] = image_cvtGray[:, w//2:]
    show_image(image_cvt, "Left Width Color and Right Width Gray")
    print("Shape of orignal image: ", image_cvt.shape)
    print("Shape of original gray: ",image_cvtGray.shape)
    print("Shape of Left: ",image_cvtGray[:, :w//2].shape)
    print("Shape of Right: ",image_cvtGray[:, w//2:].shape)
```

```python
Show_LeftorRight_Gray("LR")
```

```python
def Show_UporDown_Gray(choose):
  if choose == "U":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    image_cvt[:h//2, :, 0] = image_cvtGray[:h//2, :]
    image_cvt[:h//2, :, 1] = image_cvtGray[:h//2, :]
    image_cvt[:h//2, :, 2] = image_cvtGray[:h//2, :]
    show_image(image_cvt, "Up Width Gray and Down Width Color")
  elif choose == "D":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    image_cvt[h//2:, :, 0] = image_cvtGray[h//2:, :]
    image_cvt[h//2:, :, 1] = image_cvtGray[h//2:, :]
    image_cvt[h//2:, :, 2] = image_cvtGray[h//2:, :]
    show_image(image_cvt, "Up Width Color and Down Width Gray")
  elif choose == "UD":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    image_cvt[:h//2, :, 0] = image_cvtGray[:h//2, :]
    image_cvt[:h//2, :, 1] = image_cvtGray[:h//2, :]
    image_cvt[:h//2, :, 2] = image_cvtGray[:h//2, :]
    show_image(image_cvt, "Left Width Gray and Right Width Color")
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    image_cvt[h//2:, :, 0] = image_cvtGray[h//2:, :]
    image_cvt[h//2:, :, 1] = image_cvtGray[h//2:, :]
    image_cvt[h//2:, :, 2] = image_cvtGray[h//2:, :]
    show_image(image_cvt, "Left Width Color and Right Width Gray")
    print("Shape of orignal image: ", image_cvt.shape)
    print("Shape of original gray: ",image_cvtGray.shape)
    print("Shape of Left: ",image_cvtGray[:h//2, :].shape)
    print("Shape of Right: ",image_cvtGray[h//2:, :].shape)
```

```python
Show_UporDown_Gray("UD")
```

```python
def UporDownDiagonal_Gray(choose):
  if choose == "ULeft":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    # create mask for upper diagonal
    mask = np.zeros((h, w), dtype=np.uint8)
    for y in range(h):
        for x in range(w):
            if y < x * h / w:
                mask[y, x] = 1
    # apply image gray for upper
    image_cvt[:, :, 0] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 0])
    image_cvt[:, :, 1] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 1])
    image_cvt[:, :, 2] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 2])
    show_image(image_cvt, "Main Left Digonal Gray")
  elif choose == "DLeft":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    # create mask for upper diagonal
    mask = np.zeros((h, w), dtype=np.uint8)
    for y in range(h):
        for x in range(w):
            if y >= x * h / w:
                mask[y, x] = 1
    # apply image gray for upper
    image_cvt[:, :, 0] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 0])
    image_cvt[:, :, 1] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 1])
    image_cvt[:, :, 2] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 2])
    show_image(image_cvt, "Secondary Left Digonal Gray")
  elif choose == "UDLeft":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    # create mask for upper diagonal
    mask = np.zeros((h, w), dtype=np.uint8)
    for y in range(h):
        for x in range(w):
            if y < x * h / w:
                mask[y, x] = 1
    # apply image gray for upper
    image_cvt[:, :, 0] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 0])
    image_cvt[:, :, 1] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 1])
    image_cvt[:, :, 2] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 2])
    show_image(image_cvt, "Main Left Digonal Gray")
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    # create mask for upper diagonal
    mask = np.zeros((h, w), dtype=np.uint8)
    for y in range(h):
        for x in range(w):
            if y >= x * h / w:
                mask[y, x] = 1
    # apply image gray for upper
    image_cvt[:, :, 0] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 0])
    image_cvt[:, :, 1] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 1])
    image_cvt[:, :, 2] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 2])
    show_image(image_cvt, "Secondary Left Digonal Gray")
  elif choose == "URight":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    # create mask for upper diagonal
    mask = np.zeros((h, w), dtype=np.uint8)
    for y in range(h):
        for x in range(w):
            if y < (w - x) * h / w:
                mask[y, x] = 1
    # apply image gray for upper
    image_cvt[:, :, 0] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 0])
    image_cvt[:, :, 1] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 1])
    image_cvt[:, :, 2] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 2])
    show_image(image_cvt, "Main Right Digonal Gray")
  elif choose == "DRight":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    # create mask for upper diagonal
    mask = np.zeros((h, w), dtype=np.uint8)
    for y in range(h):
        for x in range(w):
            if y >= (w - x) * h / w:
                mask[y, x] = 1
    # apply image gray for upper
    image_cvt[:, :, 0] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 0])
    image_cvt[:, :, 1] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 1])
    image_cvt[:, :, 2] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 2])
    show_image(image_cvt, "Secondary Right Digonal Gray")
  elif choose == "UDRight":
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    # create mask for upper diagonal
    mask = np.zeros((h, w), dtype=np.uint8)
    for y in range(h):
        for x in range(w):
            if y < (w - x) * h / w:
                mask[y, x] = 1
    # apply image gray for upper
    image_cvt[:, :, 0] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 0])
    image_cvt[:, :, 1] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 1])
    image_cvt[:, :, 2] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 2])
    show_image(image_cvt, "Main Right Digonal Gray")
    image_cvt = image.copy()
    image_cvtGray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)
    # create mask for upper diagonal
    mask = np.zeros((h, w), dtype=np.uint8)
    for y in range(h):
        for x in range(w):
            if y >= (w - x) * h / w:
                mask[y, x] = 1
    # apply image gray for upper
    image_cvt[:, :, 0] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 0])
    image_cvt[:, :, 1] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 1])
    image_cvt[:, :, 2] = np.where(mask == 1, image_cvtGray, image_cvt[:, :, 2])
    show_image(image_cvt, "Secondary Right Digonal Gray")
```
