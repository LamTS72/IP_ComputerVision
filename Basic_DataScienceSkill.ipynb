# MATPLOTLIB WITH PYLOT KNOWLEDGE

```python
import matplotlib.pyplot as plt
```

```python
# create empty diagram
plt.plot()
```

```python
x = np.arange(1,5)
y = np.random.randint(1,10, 4)
print(x, y)
plt.plot(x,y, color="red")
```

```python
# Pylot API
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color="green", linestyle="dashed", label="sin(x)");
plt.plot(x, np.cos(x), color="red", label="cos(x)") # add more line in one diagram
plt.xlabel("X axis")
plt.ylabel("Y axis")
#plt.axis("equal")
plt.title("sin(x) and cos(x)")
plt.legend() # add label
plt.show()
```

```python
#Oject Oriented API
x = [1, 2, 3, 4] 
y = [4, 7, 9, 7]
print(x, y)
fig, ax = plt.subplots(figsize=(5,5)) # fig is big object, ax is sub object 
ax.plot(x, y);
ax.set(title="Sample plot",xlabel="x axis",ylabel="y axis")
```

```python
# OO API for scatter plot
rd = np.random.RandomState(0)
x = rd.randn(100)
y = rd.randn(100)

fig, ax = plt.subplots(figsize=(3,3))
plot = ax.scatter(x, y, c=rd.randn(100), cmap='viridis', alpha=0.5)
fig.colorbar(plot)
```

## 3.Bar

```python
drink_price = {
    "coke":10,
    "pepsi":12,
    "sprite":9
}

fig, ax = plt.subplots(figsize=(3,3))
ax.bar(drink_price.keys(), drink_price.values());
ax.set(title="price of drink")
```

## 4.Histogram

```python
np.random.seed(42)
student_heights = np.random.normal(170,10 ,250)
print(student_heights.shape)
```

```python
#Pylot API for histograms
plt.hist(student_heights, bins=30);
```

## 5.Subplots in a figure

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(figsize=(6,3), nrows=2, ncols=2)
x = np.array([1, 2, 3, 4])
ax1.plot(x, x/2)
ax1.plot(x, x*2)

ax2.scatter(x, x**2)

ax3.bar(drink_price.keys(), drink_price.values())

ax4.hist(student_heights)
```

```python
fig, ax = plt.subplots(figsize=(6,3), nrows=2, ncols=2)
x = np.array([1, 2, 3, 4])
ax[0,0].plot(x, x/2)
ax[0,0].plot(x, x*2)

ax[0,1].scatter(x, x**2)

ax[1,0].bar(drink_price.keys(), drink_price.values())

ax[1,1].hist(student_heights)
```

```python
df_cties = pd.read_csv("california_cities.csv")
df_cties.head()
```

```python
df_cties["area_total_km2"].sort_values(ascending=False)
```

```python
# Extract latd = V (ngang) and longd = Kinh (dc)
lat, lon = df_cties["latd"], df_cties["longd"]
print(lat[:10])
print(lon[:10])

population, area = df_cties["population_total"], df_cties["area_total_km2"]
```

```python
plt.scatter(lon, lat, c=np.log10(population), 
            cmap='viridis', s=area, linewidths=0, alpha=0.5);
plt.axis("equal")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="log10 of population")
plt.title("California City population and area")

area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s=area, label=str(area) + "km$^2$")
plt.legend()
```

# SEABORN KNOWLEDGE

```python
import seaborn as sns
%matplotlib inline

print(sns.get_dataset_names())    
```

```python
df_tips = sns.load_dataset("tips")
df_tips.head()
```

```python
sns.histplot(data=df_tips["total_bill"])
```

```python
df_tips["total_bill"].value_counts().sort_values(ascending=False)
```

```python
# KDE plot distribution
sns.kdeplot(data=df_tips["total_bill"])
```

```python
# Displot distribution = histogram + kde plot distribution
sns.displot(data=df_tips, x="total_bill", kde=True, col="time")
```

```python
# Categorical distribution with bar plot
sns.barplot(data=df_tips, x="sex", y="tip", estimator=np.mean)
```

```python
df_tips["sex"].value_counts()
```

```python
sns.countplot(data=df_tips, x="sex")
```

```python
# Box plot
sns.boxplot(data=df_tips, x="day", y="total_bill", hue="sex")
```

```python
tips_fg = sns.FacetGrid(data=df_tips, row="smoker", col="time")
tips_fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
ntips_fg = sns.FacetGrid(data=df_tips, col="sex", hue="smoker")
ntips_fg.map(sns.scatterplot, 'total_bill', 'tip', s=100, edgecolor="b", alpha=0.7)
ntips_fg.add_legend()
```
